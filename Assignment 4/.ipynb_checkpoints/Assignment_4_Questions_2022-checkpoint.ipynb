#### 1. What are the most common adverbs in the brown corpus (categories="news")? Please output the 10 most frequent ones (Please use the universal tagset).

```python
import nltk
from nltk.corpus import brown
```

```python
brown.categories()
```

```python
text=brown.words(categories='news')
```

```python
nltk_tag = nltk.pos_tag(text,tagset="universal")
```

```python
cfd_nps = nltk.ConditionalFreqDist((tag, word) for (word, tag) in nltk_tag)
```

```python
cfd_nps['ADV'].most_common(10)
```
